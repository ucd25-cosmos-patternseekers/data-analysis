```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("../data/cleaned/cleaned.csv", index_col=0)

data = data.drop(columns=['Which group is the best in Cluster 11?'])

data.columns = ['Name', "Sleep Hours", "Roommate Rating", "Food Rating", "Cow Sentiment"]

data = data.iloc[:15]

corr_matrix = data.corr(numeric_only=True)

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```
